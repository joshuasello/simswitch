```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline

import os 
from pathlib import Path
from switchsim import *


simulation_output_directory_path = Path("buck_simulation_outputs")
simulation_results_directory_path = Path("buck_simulation_results")

os.getcwd()
```

## 1. Load Simulation Config

```python
config = load_config_from_yaml(
    config_file_path="../buck_gan_simulation.yaml",
    simulation_type=SimulationType.BUCK_CONVERTER,
)
config
```

## 2. Execute or Load Simulation Outputs

```python
if not simulation_output_directory_path.exists():
    per_run_outputs = run_simulations(
        simulation_type=SimulationType.BUCK_CONVERTER,
        runs=config.runs,
        default_parameters=config.setup.default_parameters,
        output_field_mapping=config.setup.output_field_mapping,
        ltspice_executable_file_path=config.setup.ltspice_executable_file_path,
        verbose=True,
    )
    
    save_simulation_outputs(
        output_directory_path=simulation_output_directory_path,
        per_run_outputs=per_run_outputs,
    )
else:
    per_run_outputs = load_simulation_outputs(
        simulation_type=SimulationType.BUCK_CONVERTER,
        output_directory_path=simulation_output_directory_path,
    )
```

```python
per_run_outputs
```

## 3. Process Simulation Outputs

```python
per_run_results = process_simulation_outputs(
    simulation_type=SimulationType.BUCK_CONVERTER,
    per_run_outputs=per_run_outputs,
    selected_results=config.results,
)
per_run_results
```

```python
per_run_results["buck_converter_gan_gs66516t"]["switching_frequency"]
```
